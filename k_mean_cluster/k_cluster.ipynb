# K Means Cluster


```python
import seaborn
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
#artificially amke data to play around witha
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
#data[0] is the data, data[1] are the clusters 

#features = data[0]
#target = data[1]
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1]) #all blobs in first col vs all blobs in second col
```

----
#Training and getting data ourselfs

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))


ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```
