Term Frequency: Importance of the term within the document
- TF(d,t) = Number of occurances of term t in document d
- Inverse Document Fq: Importance of the term in the corpus (all docs)
    - IDF(t) = log(D/t) where
        - D = total number of documents
        - t = total number of documents with term
