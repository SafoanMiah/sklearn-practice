# Support Vector Machines

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

from sklearn.model_selection import GridSearchCV
```

```python
cancer = load_breast_cancer()
cancer.keys() #in dict format
```

```python
print(cancer['DESCR'])
```

```python
feat = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
feat.head(3)
```

```python
target = cancer['target']
target
```

.... do visualization ...

```python
X_train, X_test, y_train, y_test = train_test_split(feat, target, test_size=.3)
```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

### Creating a grid of parameters and trying out all the parameters in that grid

```python
from sklearn.model_selection import GridSearchCV
```

C: controls cost of missclassification on training data, large C = low bias, high variance, vice versa

gamma: large gamma: high bias low variance

```python
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=2)
#verbose is default 0, its just how much text output it gives
```

```python
grid.fit(X_train, y_train)

#find best parameter settings, and selects ones with best cross validation scores
```

```python
#best parameters
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.pred = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, grid.pred))
print(classification_report(y_test, grid.pred))
```
