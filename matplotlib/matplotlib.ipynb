```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#show graphs in jupyter without show()
%matplotlib inline 
```

```python
x = np.linspace(0,5,11) #0-5 wil 11 entries linearly spaced
y = x**2
```

```python
y
```

```python
# FUNCTIONAL (basic)
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

#when not in jupyter do plt.show() at end everytime
```

```python
# MULTIPLOTS
plt.subplot(2,2,1) #row, columns, plt number 2 row, 2 columns, first plot
plt.plot(x,y,'r')

plt.subplot(2,2,2) #second plot, as last num is 2
plt.plot(y,x)

plt.subplot(2,2,3) #third plot, as last num is 3
plt.plot(-x,y)
```

## OOP Plotting

```python
# CLASS OOP METHOD (better easier making figure objects, more control)
fig = plt.figure() # blank canvas

# make axis on canvas (left, bottom, width, height) 0-1
# left, bottom = start of axes
axes = fig.add_axes([0,0,1,1])

#plot axes
axes.plot(x,y)
axes.set_ylabel('Y Label')
```

```python
#mutiple axes in one canvas
fig = plt.figure()

#left, bottom, height, width, percentage (VALS BETWEEN 0-1 (0,100%))
axes1 = fig.add_axes([0.1, 0.1, 0.6, 0.6])
#cahnge left and bottom
axes2 = fig.add_axes([0.3, 0.3, 0.6, 0.6])
#change size
axes3 = fig.add_axes([0.1, 0.1, 0.3, 0.3])

axes3.plot(x,y)
axes2.plot(y,x)
axes3.set_title('S Plot')
axes2.set_title('L Plot')
```

### OOP Subplots

```python
#touple unpacking, also much faster to do it like this
fig,axes = plt.subplots(nrows=2, ncols=3) #rows and cols for multiple
plt.tight_layout() #fix overlaps

# fig is not 2D list of axes, cols or rows
print(axes) # a array of ax es so you can iterare trough it
for i in range(2):
    for j in range(3):
        axes[i,j].plot(x,y)

#or singularly, whe cols or row is 1 you can just do single bracket
axes[1,1].clear() #clear whats on it alr
axes[1,1].plot(y,x)
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
```

### Figure size, and DPI

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize=(6,3)) #you can use default
fig.tight_layout()

ax[0].plot(x,y)
ax[1].plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes((0,0,1,1)) #left, bottom, height, width

# add labels for the 
ax.plot(x,x**2, label='x^2')
ax.plot(x,x**3, label='x^3')

# add location or not
ax.legend(loc=0) #default finds best one
#ax.legend(loc='center right')
#ax.legend(loc=(0.1,0.1)) use touple, to get axes from 0 - 1
```

```python
#save figure
fig.savefig('test_plot.png', dpi=200) #resolution is DPI
```
