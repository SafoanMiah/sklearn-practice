# K-Nearest Neighbours


Classification algorithm that operates on ex:
- Data: Dogs and Horses, with Heights and Weights


Classify new datapoints within classes

#### Training: 
1. Store all the Data

#### Prediction Algirithm: 
1. Calculate the distance from x to all points on data
2. Sort the points in our data by increasing distance from x
3. Predict the majority lael of the 'k' closest points


If we chose k=3 then we look at the 3 nearest neighbours, (1 yellow, 2 purple) and predict to the class where there is more points, in this case purple.
But if we do k=6, there are 4 yellow and 2 purples, in which case we classify it in yellow


higher k, gives cleaner cutoff, at cost of mislabelling some points, more bias

#### Pros of knn:
- Very simple
- Training is trivial
- Works with any number of classes
- Easy to add more data
- Kew parameters
    - K (how many points we will look at)
    - Ditance metric (How we are mathematically defining ditance)

#### Cons of knn:
- High precision cost (worse for larg data sets)
- Not good with high dimentional data
- Caegorical features dont work well
