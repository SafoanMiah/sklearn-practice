# Logistic Regression

Method for classification (binary classification, yes no, etc.)
So far we did continous predictions, like House price but this allows us to do discrete predictions

You cannot use linear regression on binary fit



equasion for linear regression line: $$\phi(z) =\frac{1}{1+e^{-(z)}}$$


doesnt matter what value of z you put into the equasion the output will always be between 0 and 1



$$y = b_0 + b_1x$$ 
Is our linear model, in form of y=mx+b we can now take that and sub it into our z to transform it into the logistic regression line

$$\phi(z) =\frac{1}{1+e^{-(b_0 + b_1x)}}$$

The curve will be between at **0 and 1**
Thrn we can set a cutoff point at 0.5, and say anything bellow is part of class 0 and anythign above belongs to 1
