# SEABORN GRIDS
Automate subplots base don features on data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
iris.head(3)
```

```python
#empy pair grid making subplots
grid = sns.PairGrid(iris)

#grid.map(plt.scatter) #put scatter on everything

grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot)
grid.map_upper(sns.histplot)
```

```python
#map plot type and seperate results by cols and rows
grid = sns.FacetGrid(data=tips, col='time', row='smoker')
grid.map(sns.histplot, 'total_bill')
```

```python
grid = sns.FacetGrid(data=tips, col='time', row='smoker')
#if we wanted tip vs total bill you do
grid.map(sns.histplot, 'total_bill', 'tip')
```
