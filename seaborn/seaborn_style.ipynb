```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.set_style('darkgrid') #whitegrid or white

#size
plt.figure(figsize=(2,2))

#simple countplot
sns.countplot(tips, x='sex')
```

```python
sns.set_style('ticks') #adds ticks to all sides

plt.figure(figsize=(3,2))

#simple countplot
sns.countplot(tips, x='sex')

sns.despine() #by default top and right get removed, but you can specify
sns.despine(top=True, bottom=True, right=False, left=True)
```

```python
sns.set_context('notebook', font_scale=1) #context: paper notebook, talk, poster
#font scale = 3, x3 size of default

#simple countplot
sns.countplot(tips, x='sex')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex',
           palette='coolwarm')
```
