# SEABORN CATEGORICAL PLOTTING
- mainly concered with categorical columns like sex, smoker etc. (not numericals)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

sns.set_theme()
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
#like group by (x is categorical y is numerical)
#estimator: by dault is mean(), so it finds mean, but you can put other things
sns.barplot(data=tips,x='sex', y='total_bill', estimator=np.std)
```

```python
#count numer of occurunces on category
sns.countplot(data=tips, x='sex')
```

### Distribution of categorical data

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
#dots are outlies
#add another categorical using {hue} spplits even further
#here we see people on friday have higher bill if they are a smoker as box is higher
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)

#a lot more depth to it, and you can also split down the middle instead of doubling up
#violin is like a distribution while barchart is set
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
#violin + strip, no overlapping vals, clearer
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
#general plot that takes in plot types
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
```
