```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_theme(palette='coolwarm')
```

## Distribution plots
example datasets: https://github.com/mwaskom/seaborn-data

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
tips[['total_bill','tip']].corr()
```

```python
#pass in one column to quickly see distribution
sns.displot(tips['total_bill'], kde=True, bins=40)
#bins = number of bars
```

```python
#compare 2 distributions, #you can do kind: hex, reg (regression), kde, etc.
sns.jointplot(data=tips, x='total_bill', y='tip', height=4, kind='scatter') #height: size
```

```python
#compare 2 distributions
fig,ax = plt.subplots(nrows=1, ncols=3)
fig.suptitle('types of comparison graphs')

sns.regplot(ax=ax[0], data=tips, x='total_bill', y='tip')
sns.kdeplot(ax=ax[1], data=tips, x='total_bill', y='tip')
sns.histplot(ax=ax[2], data=tips, x='total_bill', y='tip', kde=True)
```

```python
#pairwise relation between numerical cols
#like joint for all num cols
sns.pairplot(tips, hue='sex', palette='coolwarm')
#hue takes in categorical column
#when you one is compared to itself it just shows distribution instead
```

```python
#shows row dashes where they are on 1 column points, hist counts those dashes
sns.rugplot(tips['total_bill'])
```
