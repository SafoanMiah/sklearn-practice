# SEABORN REGRESSION PLOT
- Using LM plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip',  hue='sex', markers=['o', 'v'])
#2 markers as male and female
#in this ex male and female have same linear fit
```

```python
#you can instead call col instead of hue to call for multiple columsn not overlapped
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', row='time')
```

```python
tips['sex_time'] = tips['sex'].astype(str) + '_' + tips['time'].astype(str)  
#categorical and str so convert both to one

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex_time')

tips
```
