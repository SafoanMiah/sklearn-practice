# PLOTLY and CUFFLINKS
- plotly : interactive map : https://plotly.com/python/getting-started/
- cufflinks : links plotly to pandas : https://github.com/santosjorge/cufflinks

In plotly you can host data vis online that you can import

```python
import numpy as np
import pandas as pd
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot #since we are using it locally
import cufflinks as cf 

cf.go_offline()
init_notebook_mode(connected=True) #connect JS to notebook
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df
```

```python
#using dict
df2 = pd.DataFrame({'Category':['A', 'B', 'C'], 'Values': [91,19,31]}) 
df2
```

```python
#df.plot() #pandas simple plot makes plot with plt


#import cufflinks as cf, cf.go_offline()
df.iplot() #IPLOT for plotly
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers' , size = 5)
#without mode plotly tries to connect lines
```

```python
df2.iplot(kind='bar', x='Category')
```

sometimes if you have have too many data points you need some kind of aggregate functions, to group by

```python
df.sum().iplot(kind='bar')
```

```python
#auto aggregations
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5],
                    'y': [10,20,30,20,10],
                    'z': [400,300,200,100,50]})
#3D plot is needed to vis it all together
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df.iplot(kind='hist', bins=30)
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```

```python
df.scatter_matrix() #like : pairplot
```
