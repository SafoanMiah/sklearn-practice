# Pricipal Component Analysis

Used to examine the interrelations in a set of variables in order to identify the underlying structure of those variables

- Regression: Line of best fit
- PCA: Orthogonal (right angle) lines in n-dimentions (as many dimentions as variables)

You want to drop components that dont explain that much of our data

---

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
feat = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
feat.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(feat)
scaled = scaler.transform(feat)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #number of components that you want to keep
```

```python
pca.fit(scaled)
```

```python
x_pca = pca.transform(scaled) #trasnform to first principle component
```

```python
scaled.shape
```

```python
x_pca.shape #reduced from 30 to 3
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target']) #all rows from col 0 agasint all rows from col 1

plt.xlabel('First Principle')
plt.xlabel('Second Principle')
```
